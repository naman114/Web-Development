```python
import math
# x' = -kx
# x = f(t)
# k = 1
# xo = 7
# to = 0
# dt = 0.1
# tf = 5

def modelODE1(k, x):
    return -k*x

k = float(input("Enter the value of k: "))
dt = float(input("Enter dt: "))
tf = float(input("Enter the final value of t: "))
to = float(input("Enter initial t: "))
xo = float(input("Enter initial x: "))

x = xo
t = to

t_val = []
x_val = []
x_actual = []

while t < tf:
    t_val.append(t)
    x_val.append(x)
    x_actual.append(xo * math.exp(-k * t))
    
    x = x + dt*modelODE1(k, x)
    t += dt
```

```python
import matplotlib.pyplot as plt

plt.plot(t_val, x_val)
plt.plot(t_val, x_actual)
```

# Second Order

```python
# x'' = -4x - 3x'

def modelODE1(t, x2):
    return x2

def modelODE2(t, x1, x2):
    return (-4*x1 - 3*x2 + 1)

k = 1
x1o = 7
x2o = 3
to = 0
dt = 0.1
tf = 5

x1 = x1o
x2 = x2o
t = to

tt = []
xx1 = []
xx2 = []

while t < tf:
    tt.append(t)
    xx1.append(x1)
    xx2.append(x2)
    
    cur_x1 = x1
    x1 += dt * modelODE1(t, x2)
    x2 += dt * modelODE2(t, cur_x1, x2)
    t += dt
```

```python
plt.plot(tt, xx1)
plt.plot(tt, xx2)
```
