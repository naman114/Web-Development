# Experiment 4: Solving ODE Using Euler and Runge Kutta Methods

## Euler Method - First Order ODE

```python
# ODE: x' = -kx

# Initial values
# k = 1
# dt = 0.1
# tf = 5
# to = 0
# xo = 7


import math
import matplotlib.pyplot as plt

# x' = -kx
def modelODE1(k, x):
    return -k*x

k = float(input("Enter the value of k: "))
dt = float(input("Enter dt: "))
tf = float(input("Enter the final value of t: "))
to = float(input("Enter initial t: "))
xo = float(input("Enter initial x: "))

x = xo
t = to

t_val = []
x_val = []
x_actual = []

while t < tf:
    t_val.append(t)
    x_val.append(x)
    x_actual.append(xo * math.exp(-k * t))
    
    x = x + dt*modelODE1(k, x)
    t += dt

    
plt.plot(t_val, x_val, label="Calculated")
plt.plot(t_val, x_actual, label="Actual")
plt.xlabel("t")
plt.ylabel("x")
plt.legend()
plt.show()
```

## Euler Method - Second Order ODE

```python
# x'' = -4x - 3x'

def modelODE1(t, x2):
    return x2

def modelODE2(t, x1, x2):
    return (-4*x1 - 3*x2 + 1)

k = 1
x1o = 7
x2o = 3
to = 0
dt = 0.1
tf = 5

x1 = x1o
x2 = x2o
t = to

tt = []
xx1 = []
xx2 = []

while t < tf:
    tt.append(t)
    xx1.append(x1)
    xx2.append(x2)
    
    cur_x1 = x1
    x1 += dt * modelODE1(t, x2)
    x2 += dt * modelODE2(t, cur_x1, x2)
    t += dt

plt.plot(tt, xx1, label="ODE1")
plt.plot(tt, xx2, label="ODE2")
plt.xlabel("t")
plt.ylabel("x")
plt.legend()
plt.show()
```

# Euler's Modified Method - Runge Kutta 2nd Order Method

```python
# ODE: x' = -kx

# Initial values
# k = 1
# dt = 0.1
# tf = 5
# to = 0
# xo = 7

import math
import matplotlib.pyplot as plt

# x' = -kx
def modelODE1(k, x):
    return -k*x

k = float(input("Enter the value of k: "))
dt = float(input("Enter dt: "))
tf = float(input("Enter the final value of t: "))
to = float(input("Enter initial t: "))
xo = float(input("Enter initial x: "))

x = xo
t = to

t_val = []
x_val = []
x_actual = []

while t < tf:
    t_val.append(t)
    x_val.append(x)
    x_actual.append(xo * math.exp(-k * t))
    
    k1 = dt*modelODE1(k, x)
    k2 = dt*modelODE1(k, x + dt)
    
    x = x + (k1 + k2)/2 
    t += dt

    
plt.plot(t_val, x_val, label="Calculated")
plt.plot(t_val, x_actual, label="Actual")
plt.xlabel("t")
plt.ylabel("x")
plt.legend()
plt.show()
```

# Runge Kutta 4th Order Method

```python
# ODE: x' = -kx

# Initial values
# k = 1
# dt = 0.1
# tf = 5
# to = 0
# xo = 7

import math
import matplotlib.pyplot as plt

# x' = -kx
def modelODE1(k, x):
    return -k*x

k = float(input("Enter the value of k: "))
dt = float(input("Enter dt: "))
tf = float(input("Enter the final value of t: "))
to = float(input("Enter initial t: "))
xo = float(input("Enter initial x: "))

x = xo
t = to

t_val = []
x_val = []
x_actual = []

while t < tf:
    t_val.append(t)
    x_val.append(x)
    x_actual.append(xo * math.exp(-k * t))
    
    k1 = dt*modelODE1(k, x)
    k2 = dt*modelODE1(k, x + dt/2)
    k3 = dt*modelODE1(k, x + dt/2)
    k4 = dt*modelODE1(k, x + dt)
    
    x = x + (k1 + 2*k2 + 2*k3 + k4)/6
    t += dt

    
plt.plot(t_val, x_val, label="Calculated")
plt.plot(t_val, x_actual, label="Actual")
plt.xlabel("t")
plt.ylabel("x")
plt.legend()
plt.show()
```
